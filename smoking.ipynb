# Clean Up Our Data

 First we will need to check for null null values in our data set. Null values often times can play a large part in data analysis and how reliable the result/conslusions are.

```python
df.isnull().values.any()
```

Since our output is false, thankfully we don't have ANY nulls. Next let's rename our columns for better readibility.  

```python
df.rename(columns = {'Data.Percentage.Female':'F_percentage'}, inplace = True)
```

```python
df.rename(columns = {'Data.Percentage.Male':'M_Percentage'}, inplace = True)
```

```python
df.rename(columns = {'Data.Smokers.Male':'M_Total'}, inplace = True)
```

```python
df.rename(columns = {'Data.Smokers.Female':'F_Total'}, inplace = True)
```

```python
df.rename(columns = {'Data.Smokers.Female':'F_Total'}, inplace = True)
```

```python
df.rename(columns = {'Data.Percentage.Total':'Tot_Percentage'}, inplace = True)
```

```python
df.rename(columns = {'Data.Percentage.Total':'Tot_Percentage'}, inplace = True)
```

```python
df.rename(columns = {'Data.Smokers.Total':'Tot_Smokers'}, inplace = True)
```

```python
df.rename(columns = {'Data.Daily cigarettes':'Daily_Cigs'}, inplace = True)
```

```python
df.head()
```

### This is much easier to read now!

### Now let's round our decimals for better accuracy we will round to 2 decimal places. 

```python
df.F_percentage = df.F_percentage.round(2)
```

```python
df.Daily_Cigs = df.Daily_Cigs.round(2)
```

```python
df.M_Percentage = df.M_Percentage.round(2)
```

```python
df.Tot_Percentage = df.Tot_Percentage.round(2)
```

```python
df.head()
```

```python
b = df.sort_values(by ='Tot_Percentage', ascending = 0)
```

```python
b.drop(b[b['Year'] != 2010].index, inplace = True)
```

```python
b
```

```python
tot = b.iloc[:3].plot.bar(x="Country", y='Tot_Percentage')
tot.set(xlabel='Country', ylabel="Smoker Population Percentage", title="Highest Smoking Percentage by Country in 2010")
```

```python
d = df.sort_values(by ='Tot_Percentage', ascending = 0)
```

```python
d.drop(d[d['Year'] != 1980].index, inplace = True)
```

```python
d
```

```python
tot = d.iloc[:3].plot.bar(x="Country", y='Tot_Percentage')
tot.set(xlabel='Country', ylabel="Smoker Population Percentage", title="Highest Smoking Percentage by Country in 19080")
```

```python
sns.set_style('whitegrid')
sns.set_context('notebook')
plt = sns.lineplot(x='Year', y='Daily_Cigs', data=df)
plt.set(xlabel='Year', ylabel="Cigarette's Smoked/Day", title="Global Daily Cigarette Usage from 1980-2010")
```
