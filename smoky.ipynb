# Smoking

### Importing libraries and reading in data file 'smoking.csv'

```python
import pandas as pd, numpy as np, seaborn as sns, matplotlib.pyplot as plt
```

```python
df = pd.read_csv('smoking.csv')
```

### Let's look at our data frame

```python
df.head()
```

# Clean Up Our Data

 First we will need to check for null null values in our data set. Null values often times can play a large part in data analysis and how reliable the result/conslusions are.

```python
df.isnull().values.any()
```

```python
df.drop(columns=['Data.Percentage.Male', 'Data.Percentage.Female', 'Data.Smokers.Female', 'Data.Smokers.Male'])
```

Since our output is false, thankfully we don't have ANY nulls. Next let's rename our columns for better readibility.  

```python
df.rename(columns = {'Data.Percentage.Total':'Tot_Percentage'}, inplace = True)
```

```python
df.rename(columns = {'Data.Smokers.Total':'Tot_Smokers'}, inplace = True)
```

```python
df.rename(columns = {'Data.Daily cigarettes':'Daily_Cigs'}, inplace = True)
```

```python
df.head()
```

### This is much easier to read now!

### Now let's round our decimals for better accuracy we will round to 2 decimal places. 

```python
df.Daily_Cigs = df.Daily_Cigs.round(2)
```

```python
df.Tot_Percentage = df.Tot_Percentage.round(2)
```

```python
df.head()
```

### Getting highest smoking countries by 'tot_percentage'


```python
high1980 = df.sort_values(by ='Tot_Percentage', ascending = 0)
```

```python
high1980.drop(high1980[high1980['Year'] != 1980].index, inplace = True)
```

```python
high1980.head()
```

```python
## creating color var for graphs! 
color = ['tab:blue', 'tab:orange', 'tab:green', 'tab:orange']
```

```python
tot = high1980.iloc[:3].plot.bar(x="Country", y='Tot_Percentage', color=color)
tot.set(xlabel='Country', ylabel="Smoker Population in %", title="Top 3 Highest Smoking Percentages by Country in 1980")
```

```python
high1980 = df.sort_values(by ='Daily_Cigs', ascending = 0)
```

```python
high1980.drop(high1980[high1980['Year'] != 1980].index, inplace = True)
```

```python
tot = high1980.iloc[:3].plot.bar(x="Country", y='Daily_Cigs', color=color)
tot.set(xlabel='Country', ylabel="Smoker Population in %", title="Top 3 Highest Smoking Percentages by Country in 1980")
```

### Now getting Highest smoking Countries in 2010 by 'tot_percentage'

### Taking a closer look at Greece over 30 years (greece = highest smoking in 2010)

```python
greeceTrend = df.sort_values(by ='Year', ascending = 0)
```

```python
greeceTrend.drop(greeceTrend[greeceTrend['Country'] != 'Greece'].index, inplace = True)
```

```python
greeceTrend.head() ## checking ascension
```

```python
gr = greeceTrend.plot(x="Year", y='Tot_Percentage', color="red")
gr.set(xlabel='Year', ylabel="Smoker Population in %", title="Smoking Percentage Trends in Greece 1980-2010")
```

```python
gr = greeceTrend.plot(x="Year", y='Daily_Cigs', color="red")
gr.set(xlabel='Year', ylabel="Smoker Cigarette Usage (# of cigs/day)", title="Smoking Percentage in Greece 1980-2010")
```
